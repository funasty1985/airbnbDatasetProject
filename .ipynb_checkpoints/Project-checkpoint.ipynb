#### loading the dataset

```python
# loading the dataset 

df = pd.read_csv('airbnb_toronto.csv')

print(f'Number of data samples : {df.shape[0]}')
print(f'\n\nndateset info :')
print(df.info())
```

```python
# drop columns more than 80% of null values
sample_num = df.shape[0]
factor = 0.8
dropping_columns = []
for ele in df.columns:
    if np.sum(df[ele].isna()) > round(sample_num*factor):
        dropping_columns.append(ele)

print("column with too much null colums : ", ", ".join(dropping_columns))

df = df.drop(columns=dropping_columns)
```

```python
# dropping rows with less than 3% of null values

less_than_three_percent_missing_values_columns = []
for c in df.columns:
    null_count = np.sum(df[c].isna())
    if null_count <= round(sample_num*0.03) and null_count > 0 :
        less_than_three_percent_missing_values_columns.append(c)
        
df = df.dropna(subset=less_than_three_percent_missing_values_columns)
```

```python
# dropping irrelevant columns 
# ref1: https://www.kaggle.com/code/jjmewtw/airbnb-cleaning-pca-clustering-estimation
# ref2: https://github.com/L-Lewis/Airbnb-neural-network-price-prediction/blob/master/Airbnb-price-prediction.ipynb
# To-do explanation
dropping_column = [
    "first_review", "last_review", "id", "host_id", 
    "license","instant_bookable", "host_name","listing_url",
    "scrape_id","last_scraped", "picture_url", "host_url",
    "host_thumbnail_url", "host_neighbourhood", "host_since", 
    "host_location","host_picture_url","host_about", "neighbourhood", 
    "neighbourhood_cleansed", "calendar_last_scraped","name","description", 
    "neighborhood_overview", "host_about", "host_verifications"]

# should we also drop host_response_time, host_response_rate, and host_acceptance_rate ?
# too much missing values and hard to interpolate


df = df.drop(columns=dropping_column)
```

#### dropping drop numeric columns with high similiarity

```python
# Before Drop
print()
# df.info()

# Select only numeric columns
numeric_df = df.select_dtypes(include=[float, int])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Set a threshold for high correlation
threshold = 0.8

# Identify pairs of highly correlated columns
high_corr_pairs = [(col, corr_matrix.columns[i]) for idx, col in enumerate(corr_matrix.columns) 
                   for i in range(idx) if abs(corr_matrix.iloc[idx, i]) > threshold]

# Create a set to hold columns to be dropped (avoid duplicates)
cols_to_drop = set()

# We'll add to this set one column from each highly correlated pair
print("similar pair columns : ")
for col1, col2 in high_corr_pairs:
    if col1 != col2:
        print(f"{col1} -- {col2}")
        cols_to_drop.add(col1)

# Drop columns
print()
print("Column Drop")
print(cols_to_drop)

df = df.drop(columns=list(cols_to_drop))
```

```python
# drop the rows without 'review_scores_rating'
df_processed = df_processed.dropna(how='any',subset='review_scores_rating')

def categorise(row):  
     if row['review_scores_rating'] >= 0 and row['review_scores_rating'] <= 1:
        return 0
     if row['review_scores_rating'] > 1 and row['review_scores_rating'] <= 2:
        return 1
     if row['review_scores_rating'] > 2 and row['review_scores_rating'] <= 3:
        return 2
     if row['review_scores_rating'] > 3 and row['review_scores_rating'] <= 4:
        return 3
     if row['review_scores_rating'] > 4:
        return 4
     
df_processed['review_class'] = df_processed.apply(lambda row: categorise(row), axis=1)
df_processed['review_class'].astype(np.int8)
df_processed = df_processed.drop(columns=['review_scores_rating'])
```

```python
def sample_from_cluster(gmm, cluster_index, n_samples):
    # Ensure valid cluster index
    if cluster_index < 0 or cluster_index >= gmm.n_components:
        raise ValueError("Invalid cluster index")

    mean = gmm.means_[cluster_index]
    cov = gmm.covariances_[cluster_index]

    # Generate samples
    return np.random.multivariate_normal(mean, cov, n_samples)

# Sample 10 points from cluster 0
cluster_samples = sample_from_cluster(gmm, cluster_index=0, n_samples=10)
```

### oversampling using GMM

```python
shortage_by_label = {}
target = y_train.value_counts().max()

for label, count in zip(y_train.value_counts().index, y_train.value_counts()):
    shortage_by_label[label] = target - count
```

```python
X_train_reduced_balanced = np.copy(X_train_reduced)
y_train_balanced = y_train

for rating_class in shortage_by_label.keys():
    shortage = shortage_by_label[rating_class]
    if shortage > 0:
        cluster_num = random.choice(clusters_by_rating[rating_class])
        new_samples = sample_from_cluster(gmm, cluster_num, shortage_by_label[rating_class])
        new_y_train = pd.Series([rating_class] * shortage_by_label[rating_class])
        y_train_balanced = pd.concat([y_train_balanced, pd.Series(new_y_train)])
        X_train_reduced_balanced = np.concatenate((X_train_reduced_balanced, new_samples))
```
